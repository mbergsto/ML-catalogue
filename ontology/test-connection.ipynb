```python
from rdflib import Graph

g = Graph()
g.parse("ML_Ontology.ttl", format="turtle")

print("Triples loaded:", len(g))

q = """
PREFIX : <http://h-da.de/ml-ontology/>
SELECT ?task ?area
WHERE {
  ?task a :ML_task ;
        :belongs_to ?area .
}
LIMIT 10
"""

for row in g.query(q):
    print(row.task, "->", row.area)
```
