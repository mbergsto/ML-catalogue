```python
# Cell 1: Imports and setup
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.grid"] = True
```

```python
df_tools_totals = pd.read_csv("Data/ml_libraries_frameworks.csv")
```

```python
# Cell 4: Bar chart  libraries/frameworks
df_lib = df_tools_totals[df_tools_totals["Group"] == "Library/Framework"].sort_values("Total", ascending=False)
plt.bar(df_lib["Query (short)"], df_lib["Total"])
plt.title("Publications by library/framework (manufacturing-related)")
plt.ylabel("Total")
plt.xticks(rotation=45, ha="right")
plt.tight_layout(); plt.show()
```

```python
# Cell 5: Bar chart  cloud platforms
df_cloud = df_tools_totals[df_tools_totals["Group"] == "Cloud Platform"].sort_values("Total", ascending=False)
plt.bar(df_cloud["Query (short)"], df_cloud["Total"])
plt.title("Publications by cloud platform (manufacturing-related)")
plt.ylabel("Total")
plt.xticks(rotation=45, ha="right")
plt.tight_layout(); plt.show()
```

```python
# Cell 6: Single-bar summary (optional)
df_sum = df_tools_totals[df_tools_totals["Group"] == "Summary"]
plt.bar(df_sum["Query (short)"], df_sum["Total"])
plt.title("Summarized query total")
plt.ylabel("Total")
plt.tight_layout(); plt.show()
```
